```python
# 1970 yılını çıkarma islemi
yuzdedoksanDf = yuzdedoksanDf[yuzdedoksanDf.year!=1970]

dataFrame = yuzdedoksanDf
dataFrame.groupby("year").mean("price")["price"]
```

```python
# transmission sütunu kaldırma  axis=1 sütun uzerinde calisir axis=0 satır üzerinde calisir
dataFrame = dataFrame.drop("transmission",axis=1)
```

```python
dataFrame.head()
```

```python
#np dizisine cevirme
y = dataFrame["price"].values

x = dataFrame.drop("price",axis=1).values

y 
```

```python
#x ve y dizilerini egiticegimiz ve test yapacağımız diziye bölme islemi
from sklearn.model_selection import train_test_split

x_train , x_test, y_train, y_test = train_test_split(x,y , test_size=0.3, random_state=10)
```

```python
len (x_train)
```

```python
# MinMaxScaler sınıfını sklearn.preprocessing'den içe aktarıyoruz.
# Bu scaler, verileri belirli bir aralığa (genellikle 0 ile 1 arasında) ölçeklendirmek için kullanılır.
from sklearn.preprocessing import MinMaxScaler

# MinMaxScaler sınıfından bir nesne oluşturuluyor.
scaler = MinMaxScaler()

# Eğitim verileri (x_train) MinMaxScaler ile öğreniliyor (fit) ve aynı zamanda dönüştürülüyor (transform).
# Veriler 0 ile 1 arasında bir aralığa ölçekleniyor.
x_train = scaler.fit_transform(x_train)

# Test verileri (x_test) için sadece dönüştürme işlemi yapılıyor.
# Fit işlemi yalnızca eğitim verileri üzerinde yapılır, test verisi bu öğrenilen dönüşümle ölçeklenir.
x_test = scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

x_train.shape
```

```python
# Boş bir Sequential model oluşturuluyor (katmanlar sırayla eklenecek)
model = Sequential()

# İlk katman ekleniyor, 12 nöronlu, ReLU aktivasyon fonksiyonuna sahip
model.add(Dense(12, activation="relu"))

# İkinci katman, yine 12 nöronlu ve ReLU aktivasyonlu
model.add(Dense(12, activation="relu"))

# Üçüncü katman, aynı şekilde 12 nöron ve ReLU aktivasyon fonksiyonu ile
model.add(Dense(12, activation="relu"))

# Dördüncü katman da aynı yapıda, 12 nöron ve ReLU aktivasyonu
model.add(Dense(12, activation="relu"))

# Son katman, çıkış katmanı, tek bir nöron (çıkış sayısı 1) - aktivasyon yok çünkü regresyon yapıyoruz
model.add(Dense(1))

# Modelin derlenmesi (compile edilmesi)
# Optimizer olarak "adam" kullanılıyor, bu genellikle daha iyi performans sağlar
# Kayıp fonksiyonu (loss) olarak "mse" (mean squared error) kullanılıyor, regresyon problemleri için uygun
model.compile(optimizer="adam", loss="mse")
```

```python
# Modeli eğitmek için .fit() metodu kullanıldı
# x_train: Eğitim verisinin girdileri (özellikler)
# y_train: Eğitim verisinin etiketleri (çıktılar)
# validation_data: Doğrulama verisi, modelin performansını değerlendirmek için (x_test, y_test)
# batch_size: Her adımda kaç verinin işleneceğini belirler (250 örnek bir seferde işlenecek)
# epochs: Eğitim veri seti üzerinden kaç kez geçileceğini (iterasyon) belirtir, burada 300 kez
model.fit(x = x_train, y = y_train , validation_data=(x_test, y_test), batch_size=250 , epochs=300) 
```

```python
# Modelin eğitim sırasında kaydedilen kayıp ve metrik sonuçlarını DataFrame'e donusturme
kayiplar = pd.DataFrame(model.history.history)  
# Kayıpların (loss) grafiksel olarak görselleştirilmesi olusur
kayiplar.plot()  
```

```python
# MSE (Mean Squared Error) ve MAE (Mean Absolute Error) hesaplamak için gerekli metrikler sklearn'den içe aktarma
from sklearn.metrics import mean_squared_error, mean_absolute_error  

# Model, test verisi olan x_test üzerinde tahmin yapıyor.
# Tahmin edilen sonuçlar 'tahminDizisi' adlı değişkene atanıyor.
tahminDizisi = model.predict(x_test)  


tahminDizisi
```

```python
mean_absolute_error(y_test,tahminDizisi)
```
